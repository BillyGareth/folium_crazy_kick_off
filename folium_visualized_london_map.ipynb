```python
import folium

london_location = [51.507351, -0.127758]

tool_tip='Oxford'
popup='Hey there'

m = folium.Map(location=london_location, zoom_start=15)
folium.Marker(
    location=[51.507351, -0.127758],
    popup='This is london_boy'
).add_to(m)
folium.CircleMarker(location=[51.507351, -0.127758],
                    radius=50,
                    fill=True,
                    color='green',
                   popup=folium.Popup('Feels_good_here')).add_to(m)
m
```
